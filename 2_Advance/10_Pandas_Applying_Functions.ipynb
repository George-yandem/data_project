```python
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df.salary_year_avg
```

```python
df_salary = df[pd.notna(df.salary_year_avg)].copy()
def porjected_salary(salary):
    return salary*1.03
```

```python
df_salary['salary_year_inflated'] = df_salary['salary_year_avg'].apply(porjected_salary)
df_salary[['salary_year_avg','salary_year_inflated']]
```

```python
df_salary['salary_year_inflated'] = porjected_salary(df_salary.salary_year_avg)
df_salary[['salary_year_avg','salary_year_inflated']]
```

```python
def porjected_salary(row):
    if 'Senior' in row['job_title_short']:
        return 1.05*row['salary_year_avg']
    else:
        return 1.03*row['salary_year_avg']
df_salary['salary_year_inflated'] = df_salary.apply(porjected_salary,axis=1)
df_salary[['job_title_short','salary_year_avg','salary_year_inflated']]
```
