```python
# List of transactions (+ for deposits, - for withdrawals)
transactions = [1200, -500, 300, -7000, 150, -50, 9000, -1000, -200, 5000]

# Threshold for large transactions
large_transaction_threshold = 5000
# Initialize totals
total_deposits = 0
total_withdrawals = 0

for num in transactions:
  transaction_type = "Deposit" if num > 0 else "Withdrawal"
  transaction_size = "(Large)" if abs(num) > large_transaction_threshold else ""
  if num > 0:
    total_deposits+=abs(num)
  else:
    total_withdrawals+=abs(num)
  print(f"{transaction_type}: ${abs(num)} {transaction_size}")
print(f"Summary:\nTotal Deposit: ${total_deposits}\nTotal Withdrawal: ${total_withdrawals}")
```

```python
def calculate_salaries(based_salaries, bonus_rates):
    total_salaries = []  # Create an empty list to store results
    for base, rate in zip(based_salaries, bonus_rates):
        total_salaries.append(base * (1 + rate))
    return total_salaries

salary_G1 = [3000, 4200, 5500, 6666]
rate_G1 = [0.30, 0.25, 0.10, 0.5]

result = calculate_salaries(salary_G1, rate_G1)
print(result)  # Output: [3900.0, 5250.0, 6050.0, 9999.0]
```

```python
def langmuir_adsorption(Qmax, K, C):
    """
    Calculate the adsorbed amount of a contaminant based on the Langmuir isotherm model.

    Parameters:
    Qmax : float : Maximum adsorption capacity (mg/g)
    K    : float : Langmuir constant (L/mg)
    C    : float : Equilibrium concentration (mg/L)

    Returns:
    float : Adsorbed amount (mg/g)
    """
    if C < 0 or Qmax <= 0 or K <= 0:
        return "Invalid input: C must be  0, Qmax and K must be > 0."

    q = (Qmax * K * C) / (1 + K * C)
    return q

# Example Usage:
Qmax = 50.0  # mg/g (hypothetical iron oxide adsorption capacity)
K = 0.2      # L/mg (Langmuir constant for Ga adsorption)
C = 10.0     # mg/L (Gallium concentration in solution)

adsorbed_amount = langmuir_adsorption(Qmax, K, C)
print(f"Adsorbed Amount: {adsorbed_amount:.2f} mg/g")
```
